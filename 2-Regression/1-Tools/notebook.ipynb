```python
X, y = datasets.load_diabetes(return_X_y=True)

# Extract blood pressure column (index 3)
X_bp = X[:, 3]
print(X_bp.shape)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Create a linear regression model and train it with our data

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
 # predict the target values for test data
y_pred = model.predict(X_test)
```

```python
# Predict values (already present, safe to call again)
y_pred = model.predict(X_test)

# Use the blood pressure feature (column index 3) for x-axis
x_bp_test = X_test[:, 3]

# Sort by the blood pressure feature and apply same ordering to predictions
idx = np.argsort(x_bp_test)
x_bp_sorted = x_bp_test[idx]
y_pred_sorted = y_pred[idx]
y_test_sorted = y_test[idx]

# Scatter plot of actual test data (use 1D blood pressure feature)
plt.scatter(x_bp_test, y_test, color='purple', alpha=0.6, label='Actual')

# Line plot of predicted values (sorted for a continuous line)
plt.plot(x_bp_sorted, y_pred_sorted, color='pink', linewidth=3, label='Predicted')

# Labels, title and legend
plt.xlabel('Blood Pressure Feature')
plt.ylabel('Disease Progression')
plt.title('Linear Regression on Diabetes Dataset')
plt.legend()
plt.grid(alpha=0.3)

# Show plot
plt.show()
```
